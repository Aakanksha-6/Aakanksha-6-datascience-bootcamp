```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix
```

advance data processing

```python
df["embarked"].fillna("S",inplace=True)
```

```python
df["deck"].fillna("Unknown", inplace=True)
```

```python
df.drop(["class","who","adult_male","deck","embark_town","alive"],axis=1,inplace=True)
```

converting categorical into numerical

```python
df["sex"]=df["sex"].map({"male":0,"female":1})
df["embarked"]=df["embarked"].map({"S":0,"C":1,"Q":2})
df["alone"]=df["alone"].map({True:1,False:0})
df.fillna(0, inplace=True)
```

first machine learning step

```python
x=df.drop("survived",axis=1) # everything after removing suvived is x
y=df["survived"] # and y will be only survived
```

train and test

```python
x_train,x_test,y_train,y_test=train_test_split(
    x,y,test_size=0.2,random_state=42   # takes random data not in sequence
)
```

```python
model=LogisticRegression(max_iter=1000)
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

confusion matrix

```python
print("Accuracy:",accuracy_score(y_test,y_pred))
```

```python
cm=confusion_matrix(y_test,y_pred)
sns.heatmap(cm,annot=True,fmt="d")
plt.show()
```

0 people notsurvied and predicted to be notsurvived(FN)  
31 people survived and predicted to be survived(TP)  
0 & 7 actual 7 died in predicted not died
