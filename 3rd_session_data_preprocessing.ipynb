```python
print(df.isnull().sum())
```

```python
#delete values from dataset
df.dropna(inplace=True)
```

```python
print(df[df['age']>30])
```

```python
print(df[df['sex']=="female"])
```

```python
print(df.groupby('sex')["survived"].mean())
```

```python
df_sorted=df.sort_values(by='age',ascending=True) #assending
print(df_sorted.head)
```

```python
#create new column
df['Family_size']=df["sibsp"]+df["parch"]
print(df.head())
```

```python
#highest fare in dataset
print(df['fare'].max())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#survival rate(percentage)
survival_rate=df['survived'].value_counts(normalize=True)*100
survival_rate.plot(kind="bar")
plt.title("Survival Percentage")
plt.ylabel("Percentage")
plt.show()
```

survival percentage-65% of people survived

```python
sns.barplot(x="sex",y="survived",data=df)
plt.title("Survival Rate by Sex")
plt.show()
```

survival rate by sex: more females survived then men

```python
sns.boxplot(x="pclass",y="fare",data=df)
plt.title("Fare Distribution by Passenger Class")
plt.show()
```

rich people i.e class 1 people paid more fare

```python
df["Family_size"]=df["sibsp"]+df["parch"]
sns.barplot(x="Family_size",y="survived",data=df)
plt.title("Survival Rate by Family Size")
plt.show()
```

```python
sns.histplot(df["age"],color="red")
plt.title("Age Distribution")
plt.show()
```

```python
sns.histplot(data=df,x="age",hue="survived",palette=["red","yellow"])
plt.title("Age Distribution by Survival")
plt.show()
```

```python
survival=df["survived"].value_counts()
plt.pie(survival,labels=["Not Survived","Survived"],autopct="%1.1f%%",colors=["blue","green"])
plt.title("Survival Percentage")
plt.show()
```

```python
sns.scatterplot(x="age",y="fare",data=df)
plt.title("Age vs Fare Scatter Plot")
plt.show()
```
