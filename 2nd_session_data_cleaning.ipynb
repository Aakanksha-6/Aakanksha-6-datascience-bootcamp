```python
import pandas as pd
```

```python
data={
    "Name":["Aman","Riya","Chetan"],
    "Marks":[23,78,45],
    "Age":[18,17,19]
}
```

```python
#dataframe is the data structure used in pandas
df=pd.DataFrame(data)
```

```python
print(df)
```

```python
print(df.columns)
```

```python
print(df.shape)
```

```python
print(df.info())
```

```python
print(df[["Name","Age"]])
```

```python
filtered=df[df["Marks"]<45]
print(filtered)
```

```python
df["Passed"]=df["Marks"]>=45
print(df)
```

```python
df_sorted=df.sort_values(by="Marks",ascending=False)
print(df_sorted)
```
